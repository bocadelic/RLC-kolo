# Redno RLC kolo

Zadatak:
1) Resiti diferencijalnu jednacinu Lq''(t)+R(t)q'(t)+q(t)/C=E(t)
2) Plotovati grafik q(t)

Priguseno oscilovanje - kada je R^2 << 4L/C

```python
#importovanje potrebnih biblioteka
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt


#definisanje parametara i funkcija
L = 1.0  # induktor
C = 1.0  # kondenzator

#funkcija za R(t) i E(t)
def R(t):
    return 0.1  

def E(t):
    return 0 

#transformisanje diferencijalne jednacine drugog reda u sistem jednacina prvog reda
def dif_transform(t, y):
    y1, y2 = y
    dy1_dt = y2
    dy2_dt = (E(t) - R(t) * y2 - y1 / C) / L
    return [dy1_dt, dy2_dt]

#zadavanje pocetnih uslova
q0 = -50.0
dq0_dt = 5.0 
y0 = [q0, dq0_dt]

#postavljanje intervala
t_span = (0, 150)
t_eval = np.linspace(t_span[0], t_span[1], 15000)

#funkcija koja resava zadati sistem diferencijalnih jednacina
sol = solve_ivp(dif_transform, t_span, y0, t_eval=t_eval)

#crtanje grafika q(t)
plt.plot(sol.t, sol.y[0], label='q(t)')
plt.xlabel('t')
plt.ylabel('Vrednost')
plt.title('Rešenje diferencijalne jednačine sa R(t)=0.1 i E(t)=0')
plt.grid(True)
plt.legend()
plt.show()
```

Resavanje diferencijalne jednacine preko Runge-Kutta 4 metode

```python
def f(t, y):
    y1, y2 = y
    dy1_dt = y2
    dy2_dt = (E(t) - R(t) * y2 - y1 / C) / L(t)
    return [dy1_dt, dy2_dt]

def Runge_Kutta(x0, y0, xn, n):
    h = (xn - x0) / n
    x = []
    y = []
    x[0] = x0
    y[0] = y0
    
    for i in range(n):
        k1 = h * (f(x[i], y[i]))
        k2 = h * (f((x[i]+h/2), (y[i]+k1/2)))
        k3 = h * (f((x[i]+h/2), (y[i]+k2/2)))
        k4 = h * (f((x[i]+h), (y[i]+k3)))
        x[i + 1] = x[i] + h
        y[i + 1] = y[i] + k1/6 + k2/3 + k3/3 + k4/6
        
    return x, y
    
```

```python
#importovanje potrebnih biblioteka
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt


#definisanje parametara i funkcija
L = 1.0  # induktor
C = 1.0  # kondenzator

#funkcija za R(t) i E(t)
def R(t):
    return 0.1  

def E(t):
    return 0 


#transformisanje diferencijalne jednacine drugog reda u sistem jednacina prvog reda
def f(t, y):
    y1, y2 = y
    dy1_dt = y2
    dy2_dt = (E(t) - R(t) * y2 - y1 / C) / L
    return [dy1_dt, dy2_dt]

#zadavanje pocetnih uslova
q0 = -50.0
dq0_dt = 5.0 
y0 = [q0, dq0_dt]

#postavljanje intervala
t_span = (0, 150)
t_eval = np.linspace(t_span[0], t_span[1], 15000)

#funkcija koja resava zadati sistem diferencijalnih jednacina
def Runge_Kutta(x0, y0, xn, n):
    h = (xn - x0) / n
    x = []
    y = []
    x.append(x0)
    y.append(y0)
    
    for i in range(n):
        #3pomonca = f(x[i], y[i])
        k1 = h * np.array(f(x[i], y[i]))
        k2 = h * np.array(f(x[i] + h / 2, y[i] + k1 / 2))
        k3 = h * np.array(f(x[i] + h / 2, y[i] + k2 / 2))
        k4 = h * np.array(f(x[i] + h, y[i] + k3))
        x.append(x[i] + h)
        y.append(y[i] + k1/6 + k2/3 + k3/3 + k4/6)
        
    return [x,y]

a = Runge_Kutta(0, y0, 150, 1000)

b = []
for item in a[1]:
    b.append(item[0])
#print(b)

def dif_transform(t, y):
    y1, y2 = y
    dy1_dt = y2
    dy2_dt = (E(t) - R(t) * y2 - y1 / C) / L
    return [dy1_dt, dy2_dt]

#zadavanje pocetnih uslova
q0 = -50.0
dq0_dt = 5.0 
y0 = [q0, dq0_dt]

#postavljanje intervala
t_span = (0, 150)
t_eval = np.linspace(t_span[0], t_span[1], 15000)

#funkcija koja resava zadati sistem diferencijalnih jednacina
sol = solve_ivp(dif_transform, t_span, y0, t_eval=t_eval)

#crtanje grafika q(t)
plt.plot(a[0],b, label='q(t)')
plt.plot(sol.t, sol.y[0], label='q(t)')
plt.xlabel('t')
plt.ylabel('Vrednost')
plt.title('Rešenje diferencijalne jednačine sa R(t)=0.1 i E(t)=0')
plt.grid(True)
plt.legend()
plt.show()

#print(a[0])
#print(';================')
#print(a[1])
```
