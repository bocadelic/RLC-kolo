# Redno RLC kolo

Zadatak:
1) Resiti diferencijalnu jednacinu Lq''(t)+R(t)q'(t)+q(t)/C=E(t)
2) Plotovati grafik q(t)

Priguseno oscilovanje - kada je R^2 << 4L/C

```python
#importovanje potrebnih biblioteka
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt


#definisanje parametara i funkcija
L = 1.0  # induktor
C = 1.0  # kondenzator

#funkcija za R(t) i E(t)
def R(t):
    return 0.1  

def E(t):
    return 0 

#transformisanje diferencijalne jednacine drugog reda u sistem jednacina prvog reda
def dif_transform(t, y):
    y1, y2 = y
    dy1_dt = y2
    dy2_dt = (E(t) - R(t) * y2 - y1 / C) / L
    return [dy1_dt, dy2_dt]

#zadavanje pocetnih uslova
q0 = -50.0
dq0_dt = 5.0 
y0 = [q0, dq0_dt]

#postavljanje intervala
t_span = (0, 150)
t_eval = np.linspace(t_span[0], t_span[1], 15000)

#funkcija koja resava zadati sistem diferencijalnih jednacina
sol = solve_ivp(dif_transform, t_span, y0, t_eval=t_eval)

#crtanje grafika q(t)
plt.plot(sol.t, sol.y[0], label='q(t)')
plt.xlabel('t')
plt.ylabel('Vrednost')
plt.title('Rešenje diferencijalne jednačine sa R(t)=1 i E(t)=0')
plt.grid(True)
plt.legend()
plt.show()
```
